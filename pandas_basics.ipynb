##### selecting rows by criteria

loc and iloc provide a consistent way to perform all the data selections.
name assigned to each column and the value of index for each row are called labels. loc works on label.
row and columns also have 0-based indeces called positions. iloc works on position.

loc and iloc allow selecting rows, columns and even individual cells within a single operation.

syntax for loc:
df.loc([row_indexer], [col indexer])

loc is an attribute of dataframe that allows row based indexing with labels.

```python
"selecting row by label:"
data=pd.read_csv('./data/artwork_data.csv', nrows=5000, index_col = 'id')
data.head(3)
```

```python
"the row's data will be returned as a series"
data.loc[1035,:]
```

```python
data.loc[1035:1037,['artist', 'title']]
```

Note : above, loc is allowing us to slice by by both row and column within a single call. The alternative without loc - data[1035:1037][['artist','title']] would involve two calls.(This is called chained indexing). For this reason, loc is preferred.

The follwoing sows that you can omit the column altoether and retrieve by row label alone

```python
data.loc[1035]
```

```python
"selecting all columns for rows satisying a condition:"
data.loc[data['acquisitionYear']==1919, :]
```

```python
data.loc[data['height'].idxmax(), ['title','medium']]
```

```python
"selecting by index position - iloc"
data.iloc[0:2, [0,1]]
```

advice : prefer iloc and iloc for data filtering...this is the most reliable way that will not clutter your code

## Preparing data



### Cleaning Data

Real world data is often not in the desired format.eg, we might find textual data in numeric columns. With this happens, pandas will use the generic object type for such a column and the data cannot be used as a number.

To have pandas use numeric type:
pd.to_numeric(data['col'])

This will fail if the column has data that cannot be converted to numeric. In such case, use the follwoing to forcefully assign NaN to bad data points:
pd.to_numeric(data['col'], errors = 'coerce')

Assigning cleaned value back to the data frame:

data.loc[:,'col'] = pd.to_numeric(data['col'], errors = 'coerce')


```python
data=pd.read_csv('./data/artwork_data.csv',nrows=5000)
data.head(3)
```

```python
data.loc[:,'height'] = pd.to_numeric(data['height'], errors='coerce')
data.loc[:,'width'] = pd.to_numeric(data['width'], errors='coerce')
```

##### Handling Missing Data

```python
df = pd.read_csv('./data/weather_m4.csv')
df.info()
```

note that the counts against all columns are not the same. where the column count does not match the number of entries, it means those columns have missing values.

```python
df[['MIN_TEMP_GROUND', 'WEATHER_CODE']].head(15)
```

Note : in pandas, null and NaN are the same

df.isnull() -> checks wehter the data for each column in a given row is null
any() -> returns true if any column has true
so, to find all columns that have any null value:

```python
df.isnull().any()
```

```python
so, to find all rows that have any null value:
```

```python
df.isnull().any(axis = 1)
```

```python
df[df.isnull().any(axis=1)]
```

we see several rows with MIN_TEMP_GROUND set to null. this is expected in this dataset. This value gets measured every 6 hours, which is why every 5th row has value and the other records have null values. Basically we have data holes and that is expected here.
Below, we verify that every 5th row has non-null value for this column and that all other rows have null values

```python
df.loc[range(5, len(df),6), 'MIN_TEMP_GROUND'].notnull().all()
```

```python
df.loc[[row for row in range(len(df)) if row not in range(5, len(df),6)], 'MIN_TEMP_GROUND'].isnull().all()
```

"alternative: here we use drop to drop all columns that have value. by default, drop does not alter the dataframe"

```python
df['MIN_TEMP_GROUND'].drop(range(5, len(df), 6)).isnull().all()
```
